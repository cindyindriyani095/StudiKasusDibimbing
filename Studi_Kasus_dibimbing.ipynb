```python
import pandas as pd

# URL dataset
data_url = "https://archive.ics.uci.edu/static/public/17/data.csv"

# Membaca dataset ke dalam DataFrame
data = pd.read_csv(data_url)

# Menampilkan 5 baris pertama
print(data.head())
```

**CLEANING DATA**

```python
# Menghapus Kolom ID

databaru = data.drop(columns=["ID"])

print(databaru)
```

```python
# Menghapus data yang bernilai NaN
databaru = databaru.dropna()
print(databaru)
```

**EKSPLORASI DATA**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Barplot Diagnosis
plt.figure(figsize=(8, 6))
sns.countplot(x="Diagnosis", data=databaru, palette="pastel")
plt.title("Distribusi Diagnosis")
plt.xlabel("Diagnosis")
plt.ylabel("Jumlah")
plt.show()
```

```python
# Distribusi untuk setiap variabel numerik
numerical_columns = databaru.select_dtypes(include=["float64", "int64"]).columns

for col in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(databaru[col], kde=True, color="red", bins=30)
    plt.title(f"Distribusi {col}")
    plt.xlabel(col)
    plt.ylabel("Frekuensi")
    plt.show()
```

```python
# Menampilkan jumlah kolom
jumlah_kolom = databaru.shape[1]
print(f"Jumlah kolom: {jumlah_kolom}")
```

```python
# Boxplot Diagnosis vs variabel numerik
for col in numerical_columns[:30]:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x="Diagnosis", y=col, data=databaru, palette="Set2")
    plt.title(f"{col} Berdasarkan Diagnosis")
    plt.xlabel("Diagnosis")
    plt.ylabel(col)
    plt.show()
```

```python
# Violin plot untuk melihat distribusi
for col in numerical_columns[:30]:
    plt.figure(figsize=(8, 6))
    sns.violinplot(x="Diagnosis", y=col, data=databaru, palette="muted")
    plt.title(f"Distribusi {col} Berdasarkan Diagnosis")
    plt.xlabel("Diagnosis")
    plt.ylabel(col)
    plt.show()
```

```python
# Menghitung distribusi kelas pada kolom 'Diagnosis'
class_distribution = databaru['Diagnosis'].value_counts()

print("Distribusi Kelas Diagnosis:")
print(class_distribution)

# Visualisasi distribusi kelas
sns.countplot(x='Diagnosis', data=databaru)
plt.title("Distribusi Kelas Diagnosis")
plt.xlabel("Diagnosis")
plt.ylabel("Jumlah")
plt.show()

# Cek proporsi kelas
class_proportion = class_distribution / len(databaru)
print("\nProporsi Kelas Diagnosis:")
print(class_proportion)
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import RandomOverSampler
from collections import Counter

# Memisahkan fitur dan target
X = databaru.drop('Diagnosis', axis=1)
y = databaru['Diagnosis']

# Oversampling dengan RandomOverSampler
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

# Menampilkan distribusi kelas setelah oversampling
print("Distribusi kelas setelah oversampling:", Counter(y_resampled))

# SMOTE untuk oversampling kelas minoritas
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Menampilkan distribusi kelas setelah SMOTE
print("Distribusi kelas setelah SMOTE:", Counter(y_resampled))
```

**KLASIFIKASI**

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Split data menjadi train dan test
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)

models = {
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "SVM": SVC(random_state=42),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=42)
}

# Melakukan klasifikasi dengan cross-validation dan evaluasi model
for model_name, model in models.items():
    print(f"\nModel: {model_name}")

    # Cross-validation untuk mengukur performa
    cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
    print(f"Cross-validation accuracy: {cv_scores.mean():.4f}  {cv_scores.std():.4f}")

    # Melatih model dengan seluruh data train
    model.fit(X_train, y_train)

    # Prediksi pada data test
    y_pred = model.predict(X_test)

    # Evaluasi pada data test
    print("\nEvaluasi Model pada Data Test:")
    print(f"Akurasi: {accuracy_score(y_test, y_pred):.4f}")
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Visualisasi confusion matrix
    plt.figure(figsize=(6, 5))
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues", xticklabels=['B', 'M'], yticklabels=['B', 'M'])
    plt.title(f'Confusion Matrix - {model_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Feature importance (untuk model yang mendukung)
    if hasattr(model, 'feature_importances_'):
        feature_importance = model.feature_importances_
        importance_df = pd.DataFrame({
            'Feature': X_train.columns,
            'Importance': feature_importance
        }).sort_values(by='Importance', ascending=False)

        print("\nFeature Importance:")
        print(importance_df)
```
